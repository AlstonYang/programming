```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([[2,2],[3,2.5],[3,9],[4,7],[5,3]])
y=np.array([1,1,2,2,2])
```

```python
plt.scatter(x[:,0],x[:,1],s=50,c=y)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()

clf.fit(x,y)
print(clf.predict([[2,2]]))
print(clf.predict(x))
```

<h3>meshgrid 產生網點</h3>

```python
X, Y = np.meshgrid(np.linspace(2,5,30), np.linspace(2,9,30))
```

```python
print(len(X))
```

<h3>ravel 變成只有一列之陣列</h3>

```python
X=X.ravel()
Y=Y.ravel()
```

<h3>zip 將兩List中element組合成各自的List</h3>

```python
datas = list(zip(X,Y))
```

```python
plt.scatter(X,Y,c=clf.predict(datas))
```
