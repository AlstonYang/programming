```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(0,5,50)
y1 = 1.2*x1 +0.8


y2 = y1 + 0.6*np.random.randn(50)
plt.scatter(x1,y2)
plt.plot(x1,y1,'r')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x2 = x1.reshape(50,1)
regr = LinearRegression()

regr.fit(x2, y2)
y3 = regr.predict(x2)

plt.scatter(x1,y2)
plt.plot(x2,y3,color='r')
plt.plot(x1,y1,color='g')
```

<h1>OverFitting</h1>

```python
from ipywidgets import interact
```

```python
x3 = np.linspace(0,5,1000)

def my_fit(n):
    y4 = 4*np.sin(n*x3)+4
    plt.scatter(x1,y2)
    plt.plot(x3,y4,color='r')
    plt.show()
```

```python
interact(my_fit, n=(1,500))
```
