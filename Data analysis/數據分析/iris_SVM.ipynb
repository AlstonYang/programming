```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
Y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(X[:,2:], Y,
                                                    test_size=0.2,
                                                   random_state = 87)

```

```python
clf = SVC()
```

```python
clf.fit(x_train,y_train)
plt.scatter(x_train[:,0],x_train[:,1],c=clf.predict(x_train))
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=clf.predict(x_test)-y_test)
```

```python
x1, y1 = np.meshgrid(np.arange(0,7,0.02), np.arange(0,3,0.02))
```

```python
z = np.c_[x1.ravel(),y1.ravel()]
result = clf.predict(z)
plt.contourf(x1,y1,result.reshape(x1.shape),cmap=plt.cm.cool,alpha=0.3)
plt.scatter(X[:,2],X[:,3],c=Y)
```
