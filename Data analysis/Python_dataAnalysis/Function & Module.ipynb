```python
def GetCrimeData(filename):
    data =[]
    with open(filename, encoding='big5') as f:
        f.readline()
        for line in f:
            data.append(line.split(','))
    return data

def GetYear(record):
    return(int(record[2][:3]))

def GetMonth(record):
    return(int(record[2][3:5]))

def GetTime(record):
    start = int(record[3][:2])
    end = int(record[3][3:])
    return(start, end)
    
def GetYearCount(data, year):
    count = 0
    for record in data:
        if GetYear(record) == year:
            count+=1
    return(count)

def GetMonthCount(data, month):
    count = 0 
    for record in data:
        if GetMonth(record)==month:
            count+=1
    return count

def GetTimeCount(data, time):
    count=0
    for record in data:
        start, end = GetTime(record)
        if start<= time < end:
            count+=1
    return count

def main():
    data = GetCrimeData('/Users/alston/bike.txt')
    GetYear(data[0])
    GetMonth(data[0])
    GetTime(data[0])
    
    year = 107
    month = 6
    time = 13
    print('%d年總案件數：%d件'%(year, GetYearCount(data, year)))
    print('%d月總案件數：%d件'%(month, GetMonthCount(data, month)))
    print('%d時總案件數：%d件'%(time, GetTimeCount(data, time)))
    
main()
```

```python
def GetGroupCount(datas):
    
    results=[]
    found = False
    for record in datas:
        start, end = bkc.GetTime(record)
        month = bkc.GetMonth(record)
        
        for result in results:
            if month == result[0][0] and start == result[0][1][0] and end == result[0][1][1]:
                result[1] +=1
                
    
        if not found:
            results.append([(month,(start, end)),1])

    return results

def MonthCount(pmhcs):
    
    counts = [[0]*24 for i in range(13)]
    
    for pmhc in pmhcs:
        counts[pmhc[0][0]][pmhc[0][1][0]-1] += pmhc[1]
        counts[pmhc[0][0]][pmhc[0][1][0]] += pmhc[1]
        
    return counts

def main():
    datas = bkc.GetCrimeData('/Users/alston/bike.txt')
    pmhcs = GetGroupCount(datas) 
    counts = MonthCount(pmhcs)
    x = []
    y = []
    s = []
    c = []
    print(counts)
    for i in range(13):
        for j in range(24):
            x.append(i)
            y.append(j)
            s.append(counts[i][j])
            c.append('red' if counts[i][j]>15 else 'blue')
            
    plt.title("Theft by month and time", fontsize=16, loc='left')
    plt.xlabel("month")
    plt.ylabel("hour")
    plt.xticks(range(1,13))
    plt.yticks(range(1,25))
    plt.scatter(x,y,s,c)
    plt.show()
main()
```
