```python
#Wiki API

def getWikiInfo(term):
    
    import requests

    url = 'https://zh.wikipedia.org/w/api.php?action=query&list=search&format=json&srsearch=%s&srlimit=500&srprop=size' %term
    
    res = requests.get(url)
    nest = res.json()
    result = []
    
    for search in nest['query']['search']:
        result.append((search['title'], search['pageid']))
        
    print(result)
    

getWikiInfo('淡江大學')
```
