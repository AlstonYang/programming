```python
from bs4 import BeautifulSoup

sp = BeautifulSoup(html, 'html.parser')
emails = re.findall(r'[\da-zA-Z_]+@[\da-zA-Z\._]+',html)
for email in emails:
    print(email)

prices = re.findall(r'[\d]+', sp.select('.price')[0].text)[0]
print(prices)

regex = re.compile(r'[\da-zA-Z\./]+\.jpg')

imgs = sp.find_all('img',{'src':regex})
for img in imgs:
    print(img['src'])
```
